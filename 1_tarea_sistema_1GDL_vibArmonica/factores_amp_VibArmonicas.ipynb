**Factores de amplificación en sistemas de 1 GDL sometidos a excitaciones armónicas**

```python
# Importación de los módulos de cálculos y visualización
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Rango de factores de amortiguación
zeta = np.array([0.05, 0.10, 0.20, 0.25, 0.50, 1])

# Rango de relación de frecuencias
r = np.linspace(0, 5, 1000)
```

*Caso 1: Vibración armónicamente excitada*

```python
# Función del factor de amplificación G
def G(r, zeta):
    return 1 / ((1-r**2)**2 + (2*zeta*r)**2)**(1/2)
```

```python
Function005 = np.zeros([1000])
Function010 = np.zeros([1000])
Function020 = np.zeros([1000])
Function025 = np.zeros([1000])
Function050 = np.zeros([1000])
Function100 = np.zeros([1000])

for i in range(0, 1000):
    Function005[i] = G(r[i], zeta[0])
    Function010[i] = G(r[i], zeta[1])
    Function020[i] = G(r[i], zeta[2])
    Function025[i] = G(r[i], zeta[3])
    Function050[i] = G(r[i], zeta[4])
    Function100[i] = G(r[i], zeta[5])
    
plt.plot(r, Function005, 'r', label='$\zeta=0.05$')
plt.plot(r, Function010, 'y', label='$\zeta=0.10$')
plt.plot(r, Function020, 'g', label='$\zeta=0.20$')
plt.plot(r, Function025, 'c', label='$\zeta=0.25$')
plt.plot(r, Function050, 'b', label='$\zeta=0.50$')
plt.plot(r, Function100, 'm', label='$\zeta=1.00$')
plt.xlim(0, 4)
plt.ylim(0, 5)
plt.xlabel("Relación de frecuencias $r$ ")
plt.ylabel("Factor de amplificación $G$")
plt.title("Vibración armónicamente excitada")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.grid()
```

*Caso 2: Vibración armónicamente excitada por desbalanceo*

```python
# Función del factor de amplificación Gdesb
def Gdesb(r, zeta):
    return r**2 / ((1-r**2)**2 + (2*zeta*r)**2)**(1/2)
```

```python
Function005 = np.zeros([1000])
Function010 = np.zeros([1000])
Function020 = np.zeros([1000])
Function025 = np.zeros([1000])
Function050 = np.zeros([1000])
Function100 = np.zeros([1000])

for i in range(0, 1000):
    Function005[i] = Gdesb(r[i], zeta[0])
    Function010[i] = Gdesb(r[i], zeta[1])
    Function020[i] = Gdesb(r[i], zeta[2])
    Function025[i] = Gdesb(r[i], zeta[3])
    Function050[i] = Gdesb(r[i], zeta[4])
    Function100[i] = Gdesb(r[i], zeta[5])
    
plt.plot(r, Function005, 'r', label='$\zeta=0.05$')
plt.plot(r, Function010, 'y', label='$\zeta=0.10$')
plt.plot(r, Function020, 'g', label='$\zeta=0.20$')
plt.plot(r, Function025, 'c', label='$\zeta=0.25$')
plt.plot(r, Function050, 'b', label='$\zeta=0.50$')
plt.plot(r, Function100, 'm', label='$\zeta=1.00$')
plt.xlim(0, 4)
plt.ylim(0, 5)
plt.xlabel("Relación de frecuencias $r$ ")
plt.ylabel("Factor de amplificación $G$")
plt.title("Vibración armónicamente excitada por desbalanceo")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.grid()
```
