**Tarea 1: Errores de aproximación sobre frecuencias para sistemas de 1 GDL sometidos a excitaciones armónicas**

```python
# Importación de los módulos de cálculos y visualización
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Rango del factor de amortiguación
ceda = np.linspace(0, 1, 1000)
```

*Caso 1: Error de aproximación para frecuencia amortiguada igual a frecuencia natural*

```python
# Expresión del error para wd=wn
def Errorwd(ceda):
    return abs(np.sqrt(1 - ceda**2) - 1) * 100
```

```python
Ewd = np.zeros([1000])
for i in range(0, 1000):
    Ewd[i] = Errorwd(ceda[i])

# Generación de gráfica
plt.plot(ceda, Ewd, '-b')
plt.xlim(0, 1)
plt.ylim(0, 100)
plt.xlabel("Factor de amortiguación $\zeta$ ")
plt.ylabel("Error [%]")
plt.title("Error de aproximación $\omega_d$ = $\omega_n$")
```

*Caso 2: Error de aproximación para frecuencia crítica igual a frecuencia natural*

```python
# Expresión del error para wc=wn
def Errorwc(ceda):
    return abs(np.sqrt(1 - 2*ceda**2) - 1) * 100
```

```python
Ewc = np.zeros([1000])
for i in range(0, 1000):
    Ewc[i] = Errorwc(ceda[i])

# Generación de gráfica
plt.plot(ceda, Ewc, '-r')
plt.xlim(0, 1)
plt.ylim(0, 100)
plt.xlabel("Factor de amortiguación $\zeta$ ")
plt.ylabel("Error [%]")
plt.title("Error de aproximación $\omega_c$ = $\omega_n$")
```

*Caso 3: Error de aproximación para frecuencia crítica por desbalanceo igual a frecuencia natural*

```python
# Expresión del error para wc_desb=wn
def Errorwc_desb(ceda):
    return abs((1/np.sqrt(1 - 2*ceda**2)) - 1) * 100
```

```python
Ewc_desb = np.zeros([1000])
for i in range(0, 1000):
    Ewc_desb[i] = Errorwc_desb(ceda[i])

# Generación de gráfica
plt.plot(ceda, Ewc_desb, '-g')
plt.xlim(0, 1)
plt.ylim(0, 100)
plt.xlabel("Factor de amortiguación $\zeta$ ")
plt.ylabel("Error [%]")
plt.title("Error de aproximación para $\omega_{c, desb}$ = $\omega_n$")
```

*Gráfico comparativo*

```python
plt.plot(ceda, Ewd, 'b', label='$\omega_d$ = $\omega_n$')
plt.plot(ceda, Ewc, 'r', label='$\omega_c$ = $\omega_n$')
plt.plot(ceda, Ewc_desb, 'g', label='$\omega_{c, desb}$ = $\omega_n$')
plt.xlim(0, 1)
plt.ylim(0, 100)
plt.xlabel("Factor de amortiguación $\zeta$ ")
plt.ylabel("Error [%]")
plt.title("Error de aproximación sobre frecuencias")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.grid()
```
